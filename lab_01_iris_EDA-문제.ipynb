####  문제 2)  모든 수치형 변수의 평균과 중앙값, 표준편차, 4분위 수를 출력하세요

```python
for col in iris.columns:
    try:
        print('%s의 평균 : %f' %(col, iris[col].mean()))
        print('%s의 중앙값 : %f' %(col, iris[col].median()))
        print('%s의 표준편차 : %f' %(col, iris[col].std()))
        print('%s의 1분위 수 : %f' %(col, iris[col].quantile(0)))
        print('%s의 2분위 수 : %f' %(col, iris[col].quantile(0.25)))
        print('%s의 3분위 수 : %f' %(col, iris[col].quantile(0.5)))
        print('%s의 4분위 수 : %f' %(col, iris[col].quantile(0.75)))
        print('%s의 5분위 수 : %f' %(col, iris[col].quantile(1)))
        print('======================')
    except:
        pass
```

#### 문제 3)  꽃받침길이 분포  ("SepalLength")

```python
plt.xlabel('길이')
plt.ylabel('개수')
plt.hist(iris['sepal.length'], bins = 20)
plt.title('꽃받침 길이 분호')
plt.show()
```

#### 문제 4)  붓꽃 품종별 꽃받침길이 분포  

```python
import seaborn as sns
sns.boxplot(x = 'variety', y = 'sepal.length', data = iris)
plt.title('붓꽃 품종별 꽃받침 길이')
plt.xlabel('품종')
plt.ylabel('꽃받침 길이')
plt.show()
```

#### 문제 8)  전체 붓꽃 꽃받침폭 분포 히스토그램  

```python
import seaborn as sns
plt.hist(iris['sepal.width'], histtype='stepfilled')
plt.xlabel('꽃 받침 폭')
plt.ylabel('도수')
plt.title('꽃 받침 폭 분포')
plt.show()
```

#### 문제 9)  전체 붓꽃 품종별 꽃받침폭 분포 히스토그램

```python
idx = 0
for name in set(iris['variety']):
    idx+=1
    plt.subplot(3,1,idx)
    plt.hist(iris.loc[iris['variety']== name, ('petal.width')])
    plt.title('%s 꽃 받침폭 분포' % (name))
    plt.xlabel('꽃 받침폭')
    plt.ylabel('개수')
    plt.show()
```
