#### 산점도 그래프

```python
import matplotlib
from matplotlib import font_manager,rc
import matplotlib.pyplot as plt
%matplotlib inline

#  #한글 폰트 등록
font_location = "c:/Windows/fonts/malgun.ttf" # Windows
# font_location = "/System/Library/fonts/AppleSDGothicNeo.ttc"  # Mac OS
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 데이터 준비
import pandas as pd

students = pd.read_csv("data/students.csv")
students.head()
```

```python
plt.plot( students["english"], students["math"], 'bo')
plt.xlabel('english')
plt.ylabel('math')
plt.title(" 영어접수와 수학점수 산점도 그래프")
plt.show() 
```

```python
import seaborn as sns
# 데이터 준비
iris = sns.load_dataset("iris")  # seaborn 팩키지의 샘플 데이터
iris.head()
```

```python
plt.plot( iris["sepal_width"], iris["petal_width"], '^')
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.title("붓꽃의 sepal_width와 petal_width 산점도 그래프")
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
             y_vars=["petal_width", "petal_length"])
plt.show()
```
